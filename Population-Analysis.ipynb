```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
years = df1.columns[1:]
total_values=df1[years].sum()
plt.figure(figsize=(30, 30))
plt.barh(years,total_values,edgecolor="white")
plt.xlabel("Total values", size= 20)
plt.ylabel("Years", size= 20)
plt.title("Total Values Per Year", size= 30)
plt.show()
```

```python
top_20_countries = df1.nlargest(20, '1960')

plt.figure(figsize=(12, 10))
plt.bar(top_20_countries['Country Name'], top_20_countries['1960'],color="blue")
plt.xlabel('Country')
plt.ylabel('Population')
plt.title('Top 20 Most Populated Countries in 1960')
plt.xticks(rotation=60)
plt.show()
```

```python
top_20_countries = df1.nlargest(20, '2023')

plt.figure(figsize=(12, 10))
plt.bar(top_20_countries['Country Name'], top_20_countries['2023'],color="purple")
plt.xlabel('Country')
plt.ylabel('Population')
plt.title('Top 20 Most Populated Countries in 2023')
plt.xticks(rotation=60)
plt.show()
```
